# Problem 2 Part 2

```python
import numpy as np
import mltools as ml
import matplotlib.pyplot as plt

iris = np.genfromtxt("C:/Users/Angelo/Downloads/178-hw1-code/data/iris.txt"
                     ,delimiter=None) # load the data
Y = iris[:,-1]
X = iris[:,0:2]    
K = [1,2,5,10,50,100,200];
errTrain = [None]*len(K) # (preallocate storage for training error)
errVal = [None]*len(K)

np.random.seed(0) # set the random number seed
X,Y = ml.shuffleData(X,Y); # shuffle data randomly
Xtr,Xva,Ytr,Yva = ml.splitData(X,Y, 0.75); # split data into 75/25 train/validation

for i,k in enumerate(K):
    learner = ml.knn.knnClassify(Xtr, Ytr, k)
    Yhat = learner.predict(Xtr)
    errTrain[i] = sum([1 if Yhat[j] != Ytr[j] 
                       else 0 for j in range(len(Ytr))]) / len(Yhat)
    
    Yhat = learner.predict(Xva)
    errVal[i] = sum([1 if Yhat[j] != Yva[j] 
                     else 0 for j in range(len(Yva))]) / len(Yhat)

plt.semilogx(K, errTrain, color='red')
plt.semilogx(K, errVal, color='green')
plt.show()
# Recommend K = 5 because training error is minimal and validation is at a local minima
```
